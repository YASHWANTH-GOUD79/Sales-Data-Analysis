# Sales Data Analysis

This notebook analyzes sales data using Python.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sales.csv')
df.head()
```

```python
df.groupby('region')['sales'].sum()
```

```python
df.groupby('region')['sales'].sum().plot(kind='bar')
plt.title('Sales by Region')
plt.show()
```
